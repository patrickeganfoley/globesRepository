#  Cool Sets of Three

```python
rotx1 = rotationFromXYZ(-np.pi/3, 0.0, 0.0)
rotx2 = rotationFromXYZ(np.pi/3, 0.0, 0.0)

roty1 = rotationFromXYZ(0.0, -np.pi/3, 0.0)
roty2 = rotationFromXYZ(0.0, np.pi/3, 0.0)

rotz1 = rotationFromXYZ(0.0, 0.0, -np.pi/3)
rotz2 = rotationFromXYZ(0.0, 0.0, np.pi/3)
```

```python
map_x1 = transform_map(bigmap, rotx1, debug = False)
map_x2 = transform_map(bigmap, rotx2, debug = False)

map_y1 = transform_map(bigmap, roty1, debug = False)
map_y2 = transform_map(bigmap, roty2, debug = False)

map_z1 = transform_map(bigmap, rotz1, debug = False)
map_z2 = transform_map(bigmap, rotz2, debug = False)
```

```python
misc.imsave('/Users/pfoley/map_x1.jpg', map_x1)
misc.imsave('/Users/pfoley/map_x2.jpg', map_x2)

misc.imsave('/Users/pfoley/map_y1.jpg', map_y1)
misc.imsave('/Users/pfoley/map_y2.jpg', map_y2)

misc.imsave('/Users/pfoley/map_z1.jpg', map_z1)
misc.imsave('/Users/pfoley/map_z2.jpg', map_z2)
```
